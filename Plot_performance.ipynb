```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def moving_average(values, window):
    weights = np.repeat(1.0, window) / window
    return np.convolve(values, weights, 'valid')
```

# -------------------------------------------------------------------------------------

```python
scores1 = np.loadtxt('losses/RL_scores_1_.txt', delimiter=',')
scores2 = np.loadtxt('losses/RL_scores_2_.txt', delimiter=',')
scores3 = np.loadtxt('losses/RL_scores_3_.txt', delimiter=',')
scores4 = np.loadtxt('losses/RL_scores_4_.txt', delimiter=',')
```

```python
s_scores1 =  moving_average(scores1, 20)
s_scores2 =  moving_average(scores2, 20)
s_scores3 =  moving_average(scores3, 20)
s_scores4 =  moving_average(scores4, 20)
```

```python
plt.plot(s_scores1, label='scores1')
plt.plot(s_scores2, label='scores2')
plt.plot(s_scores3, label='scores3')
plt.plot(s_scores4, label='scores4')

plt.legend()
plt.savefig('gen_smis/RL_all_curves.png', dpi=300)
```
